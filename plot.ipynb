```python
import numpy as np
import matplotlib.pyplot as plt

# Input sizes
sizes = np.array([1, 10, 20, 40, 80])

# Execution times
time_1node = np.array([15337, 17059, 18236, 21990, 29200])
time_2nodes = np.array([15384, 17570, 18043, 22237, 27041])
time_3nodes = np.array([14346, 16304, 19037, 20192, 27439])
time_4nodes = np.array([14609, 16307, 17711, 21711, 26907])

# Plot
plt.figure(figsize=(8, 6))

plt.plot(sizes, time_1node, marker='o', label='1 node')
plt.plot(sizes, time_2nodes, marker='o', label='2 nodes')
plt.plot(sizes, time_3nodes, marker='o', label='3 nodes')
plt.plot(sizes, time_4nodes, marker='o', label='4 nodes')

plt.xlabel("Input size (MB)")
plt.ylabel("Execution time (ms)")
plt.title("Execution Time vs Input Size for Different Node Counts")
plt.legend()
plt.grid(True)

plt.show()
```
